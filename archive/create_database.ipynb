```python
Path2 = []
for x in fileList:
    if x.startswith("2"):
        Path2.append(path + "/" + x + "/")
print(Path2)
```

```python
questionSheets = []
for i in Path2:
    filelist2 = os.listdir(i)
    for j in filelist2:
        if j.endswith(".xlsx"):
            questionSheets.append(i + j)

questionSheets
```

```python
year = questionSheets[1].split("/")[-2]
df = pd.read_excel(questionSheets[1], sheet_name="Round 1")
df["year"] = year
df.head()
```

```python
pd.ExcelFile(questionSheets[1]).sheet_names
```

```python
# to do
# For each sheet, find the names of the sheets and store in them a variable.
# Then first load sheet one in a dataframe and then iterate through the other sheets and append them to the dataframe
all_df = []

for questionSheet in questionSheets:
    sheetNames = pd.ExcelFile(questionSheet).sheet_names
    year = questionSheet.split("/")[-2]
    for sheetName in sheetNames:
        df = pd.read_excel(questionSheet, sheet_name=sheetName)
        df["year"] = year
        df["round"] = sheetName
        all_df.append(df)
```

```python
allQuestions = pd.concat(all_df)
allQuestions.shape
```

```python
allQuestions.sample(5)
```

```python
columnsToDrop = [
    "Unnamed: 11",
    "Unnamed: 12",
    "Unnamed: 13",
    "Unnamed: 14",
    "Unnamed: 15",
]

# drop columnsToDrop
allQuestions = allQuestions.drop(columnsToDrop, axis=1)
allQuestions.sample(5)
```

```python
# adding the columns necessary for creating flashcards
card_columns = {
    "due": None,
    "stability": 0,
    "difficulty": 0,
    "elapsed_days": 0,
    "scheduled_days": 0,
    "reps": 0,
    "lapses": 0,
    "state": None,
}

allQuestions = allQuestions.assign(**card_columns)

# renaming S/N to Question_number
allQuestions.rename(columns={"S/N": "Question_number"}, inplace=True)

allQuestions.sample(5)
```
